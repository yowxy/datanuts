```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from math import sqrt
```

```python
# Load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Display data info
print("Train Data Info:")
print(train.info())
print("\nTest Data Info:")
print(test.info())
```

```python
# Convert date to datetime and extract useful features
def process_dates(df):
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['day_of_week'] = df['date'].dt.dayofweek
    df['is_weekend'] = df['date'].dt.dayofweek >= 5
    return df.drop('date', axis=1)
```

```python
# Handle categorical columns properly
def encode_categorical(df, columns):
    for col in columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
    return df
```

```python

categorical_cols = ['cluster_id']  # Add other categorical columns if present
train = encode_categorical(train, categorical_cols)
test = encode_categorical(test, categorical_cols)
```

```python
def weighted_rmse(y_true, y_pred, cluster_weights):
    errors = (y_true - y_pred)**2
    weighted_errors = errors * cluster_weights
    return np.sqrt(np.mean(weighted_errors))
```

```python
def weighted_rmse(y_true, y_pred, cluster_weights):
    errors = (y_true - y_pred)**2
    weighted_errors = errors * cluster_weights
    return np.sqrt(np.mean(weighted_errors))

cluster_models = {}
rmse_results = {}
all_y_true = []
all_y_pred = []
all_weights = []

mean_consumption = train.groupby('cluster_id')['electricity_consumption'].mean()

for cluster in train['cluster_id'].unique():
    df_cluster = train[train['cluster_id'] == cluster]
    
    # Define features and target
    X = df_cluster.drop(columns=[col for col in ['electricity_consumption', 'cluster_id', 'ID', 'date'] if col in df_cluster.columns])
    y = df_cluster['electricity_consumption']
    
    # Train model
    model = RandomForestRegressor()
    model.fit(X, y)
    cluster_models[cluster] = model

    # Predict
    y_pred = model.predict(X)
    
    # Store for weighted RMSE
    all_y_true.extend(y.values)
    all_y_pred.extend(y_pred)
    all_weights.extend([mean_consumption[cluster]] * len(y))
    
    # Individual RMSE
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    rmse_results[cluster] = rmse
    print(f"RMSE Cluster {cluster}: {rmse:.2f}")

# Compute overall weighted RMSE
w_rmse = weighted_rmse(np.array(all_y_true), np.array(all_y_pred), np.array(all_weights))
print(f"\nWeighted RMSE (based on cluster mean consumption): {w_rmse:.2f}")
```
